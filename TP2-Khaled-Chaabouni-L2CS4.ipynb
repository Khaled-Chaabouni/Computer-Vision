```python
#Importing libraries :
%matplotlib inline 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
from matplotlib import pyplot, transforms
import cv2 as cv
```

```python
#Loading .jpg image into img :
img= cv.imread('Test.JPG')
```

```python
#Printing img matrix :
print(img)
```

```python
#Printing img with BGR Scheme (Pyplot default color system) :
plt.imshow(img)
```

```python
#Reversing BGR to RGB :
img2=img[:,:,::-1]
plt.imshow(img2)
```

```python
#Printing image shapes :
print(img2.shape)
print("Height of Image :",(img2.shape[0]),"pixels")
print("Width of Image :",(img2.shape[1]),"pixels")
```

```python
#Printing & Comparing pixels of 2 images :
print(img2[150,300])
print(img[150,200])
```

```python
#Adding Vertical/Horizontal lines :
for i in range(50,img2.shape[0]-50): 
    img2[i,75:100] = [255, 255, 255] 
    img2[(i+10*i)::] = [25, 50, 75]
    img2[i,100:200] = [50, 1, 300] 
img2[25:75,25:375] = [100, 100, 10]
img2[475:500,25:375] = [255, 0, 0]
plt.imshow(img2)
```

```python
#Rotating picture with 25 :
img3=cv.imread('Test.JPG')
tr = transforms.Affine2D().rotate_deg(25)
plt.imshow(img3,transform=tr)
```

```python
#Modifying image colors and saving it with cv.imwrite() :
image = cv.imread('Test.JPG', 0)
plt.imshow(image)
cv.imwrite('shadesofgrey.jpg',image)
```
